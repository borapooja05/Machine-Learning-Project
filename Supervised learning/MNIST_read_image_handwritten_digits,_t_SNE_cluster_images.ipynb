```python
from PIL import Image
import numpy as np

# Open the image file
image_path = '/content/Zero_full (1).jpg'  # Replace with your image file path
img = Image.open(image_path)

# Convert the image to a NumPy array
# For a color image (RGB), this will result in a 3D array (height, width, channels)
# For a grayscale image, it will be a 2D array (height, width)
pixel_data = np.asarray(img)
```

Here we read just one image . We could read all images and get their pixel values -store the number and pixel tuple as a dataframe to be used. Finding an effective way for loading and transforming 70K images could be a challenge. For now, using the dataset directly which already has this information

```python
from sklearn.datasets import fetch_openml
#this dataset gives pixel info for each image of a handwritten digit.
mnist=fetch_openml('mnist_784',version=1,return_X_y=True,as_frame=False) #returns a tuple
#70K records total
```

```python
#length of tuple is 2 -first part is the image pixel information,
#2nd part is the actual digit ir corresponds to
mnist
```

```python
mnist[0] #list of all 70K images' pixel information
```

```python
mnist[1] #digits in the image
```

```python
mnist[0][0] #pixel of the first image. Images seem to be in greyscale as single digits provided
```

```python
mnist[1][0] #first digit identified
```

Use t-SNE to see if digits can be distinguished

```python
X=mnist[0][:10000] #take first 1000 samples of pixel
y=mnist[1][:10000] # take first 1000 samples of digits
```

```python
X_std = X/255 #standardize - 255 max value so divide it by 255
```

```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

tsne=TSNE(perplexity=50,n_components=2,init='random')
X_embedded=tsne.fit_transform(X_std)
```

```python
X_embedded
```

```python
sns.scatterplot(x=X_embedded[:,0],y=X_embedded[:,1],hue=y)
```

Digits can be distinguished with some overlap
