Before deciding on removal/imputation technique- find correlation of individual variables with target variable. If correlation is high, keep the variable and think about imputation technique

If correlation is very low, we can consider dropping the whole column

if sample size is large and missing records is very small percentage like 5-10% then entire row could be deleted.



```python
#example of finding correlation among numerical cols - all. Similarly we can do for other categorical cols
def pairwise_correlation(df):
    """Compute pairwise correlations."""
    correlations = {}
    for col1 in df.columns:
        for col2 in df.columns:
            if col1 != col2:
                # Drop rows where either value is missing
                valid_data = df[[col1, col2]].dropna()
                if not valid_data.empty:
                    corr = valid_data.corr().iloc[0, 1]
                    correlations[f'{col1}-{col2}'] = corr
    return correlations

correlations = pairwise_correlation(df[['Age','Survived','Fare','Pclass','Parch','SibSp']])
print(correlations)
```
